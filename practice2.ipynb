```python
# Even numbers
for i in range(1, 21):
    if i % 2 == 0:
        print(i)

# Sum 1 to 100
total = 0
for i in range(1, 101):
    total += i
print("Sum:", total)

# Factorial
num = 5
fact = 1
for i in range(1, num+1):
    fact *= i
print("Factorial:", fact)
```
